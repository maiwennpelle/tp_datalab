```python
def crible_eratosthene(n) -> list[bool]:
    """Génère une liste de booléens où True indique qu'un nombre est premier."""

    est_premier = [True] * (n + 1)
    est_premier[0] = est_premier[1] = False  # 0 et 1 ne sont pas premiers

    for i in range(2, int(n**0.5) + 1):
        if est_premier[i]:
            for multiple in range(i * i, n + 1, i):
                est_premier[multiple] = False

    return est_premier
```

```python
crible_eratosthene(10)
```

```python
def nb_premiers(n,m):
    l=crible_eratosthene(m)[n:]
    c=0
    for i in l:
        if(i==True):
            c=c+1
    return c
```

```python
print(nb_premiers(10000,100000))
```

```python
import duckdb

duckdb.sql("""
CREATE OR REPLACE VIEW commune AS
FROM 'https://www.insee.fr/fr/statistiques/fichier/7766585/v_commune_2024.csv';
""")
```

```python
duckdb.sql("""
SELECT *
  FROM commune
""")
```

```python
import pandas as pd

duckdb.sql("""
SELECT *
  FROM commune
""").to_df()
```

```python
duckdb.sql("""
SELECT dep,reg
  FROM commune
  WHERE NCC='MAMOUDZOU'
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(*)
  FROM commune
  WHERE DEP='973'
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(*)
  FROM commune
  WHERE NCC LIKE '%L%L%L%L%' AND (DEP='02' OR DEP='60' OR DEP='80')
""").to_df()
```

```python
duckdb.sql("""
SELECT DEP,count(NCC)
  FROM commune
  GROUP BY DEP
  ORDER BY count(NCC) DESC
  LIMIT 20
""").to_df()
```

```python
duckdb.sql("""
SELECT DEP,NCC,len(NCC)
  FROM commune
  WHERE len(NCC)=(SELECT max(len(NCC)) FROM commune)
""").to_df()
```
